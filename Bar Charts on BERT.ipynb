```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
categories = ['Below Average', 'Above Average']
sub_categories = ['Negative', 'Positive']

# Assuming these are the probabilities or frequencies you want to plot
# For Below Average: Negative Reviews, Positive Reviews
# For Above Average: Negative Reviews, Positive Reviews
values = np.array([[0.2, 0.8], [0.1, 0.9]])  # Example values

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, values[:, 0], width, label='Negative Reviews', color='r')
rects2 = ax.bar(x + width/2, values[:, 1], width, label='Positive Reviews', color='b')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and review sentiment')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Function to autolabel bars (optional)
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
